```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import warnings

```

```python
print("\n......Dataset.......\n")
ppr=pd.read_csv('E:/Downloads/pizza_v2.csv')
```

```python
print('\n......Information about data.......\n')
ppr.info()
```

```python
print('\n......Checking the type.......\n')
ppr.dtypes.to_frame()
```

```python
print('\n......Removing Rp from price_rupiah column and inch from diameter column.......\n')
ppr['price_rupiah'] = ppr['price_rupiah'].str.replace('Rp','').str.replace(',','').astype('float64')
ppr['diameter'] = ppr['diameter'].str.replace('inch','').str.replace(',','').astype('float64')
ppr.loc[:,['price_rupiah','diameter']]
```

```python
print('\n.....Again check the type....\n')
ppr.dtypes.to_frame()
```

```python
print('\n......Structure of data.......\n')
ppr.T
```

```python
print('\n......Visualization......\n')
print('\n.......Extra sauce Barplot......\n')
plt.figure(figsize=(16,5))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='extra_sauce')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n.....Extra Cheese Barplot......\n')
plt.figure(figsize=(16,5))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='extra_cheese')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n......Extra Mushrooms........\n')
plt.figure(figsize=(16,5))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='extra_mushrooms')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n........Pizza Variant Barplot.......\n')
plt.figure(figsize=(16,16))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='variant')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n........Pizza Topping Barplot.......\n')
plt.figure(figsize=(16,12))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='topping')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n........Pizza Size Barplot.......\n')
plt.figure(figsize=(16,12))
sns.barplot(x='company',y='price_rupiah',data=ppr,hue='size')
plt.ylim(0,250000)
plt.show()
```

```python
print('\n........Company with their price........\n')
plt.figure(figsize=(16,5))
sns.swarmplot(x='company',y='price_rupiah',data=ppr)
plt.ylim(0,260000)
plt.show()
```

```python
print('\n........Company with their diameter........\n')
plt.figure(figsize=(16,5))
sns.swarmplot(x='company',y='diameter',data=ppr)
plt.ylim(0,30)
plt.show()
```

```python
print('\n........company with their price and diameter........\n')
pizza_fg=sns.FacetGrid(ppr,col='company',col_wrap=5,height=5)
pizza_fg.map(plt.plot,'price_rupiah','diameter',marker='.')
plt.show()
```
