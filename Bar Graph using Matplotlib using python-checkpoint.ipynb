```python
import matplotlib.pyplot as pyplot
# set up the data
labels = ('Cloud Computing','Software Quality','Simulation & Modeling')
index = (1,2,3) # provides location
sizes = [70, 30, 20] # provides size
# set up the bar data
pyplot.bar(index,sizes,tick_label=labels)
# configuring the layout
pyplot.ylabel('Usage')
pyplot.xlabel('Subjects scope')
# display chart
pyplot.show()

```
